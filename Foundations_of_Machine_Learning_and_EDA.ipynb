1. What is the difference between AI, ML, DL, and Data Science? Provide a brief explanation of each.

-  Artificial Intelligence (AI)
   - AI is the broadest concept. It refers to machines designed to simulate human intelligence, such as reasoning, problem-solving, decision-making, and language understanding.
   - Example: Chatbots, game-playing systems (Chess, Go), virtual assistants.
   - Scope: Broadest field; aims to create machines that can perform tasks requiring human intelligence.
   - Techniques: Rule-based systems, search algorithms, expert systems, knowledge representation, ML and DL.
   - Applications: Chatbots, robotics, game playing, decision-support systems, autonomous systems.

- Machine Learning (ML)
   - ML is a subset of AI. It focuses on systems that learn from data and improve their performance without being explicitly programmed.
   - Example: Spam email detection, recommendation systems (Netflix, Amazon).
   - Scope: Subfield of AI focused on enabling systems to learn from data.
   - Techniques: Supervised learning, unsupervised learning, reinforcement learning, regression, classification, clustering.
   - Applications: Spam filtering, recommendation systems, fraud detection, predictive analytics.

- Deep Learning (DL)
   - DL is a subset of Machine Learning. It uses neural networks with multiple layers to automatically learn complex patterns from large amounts of data.
   - Example: Image recognition, speech recognition, self-driving cars.
   - Scope: Subfield of ML dealing with complex data and large-scale learning.
   - Techniques: Artificial neural networks, convolutional neural networks (CNNs), recurrent neural networks (RNNs), transformers.
   - Applications: Image recognition, speech recognition, natural language processing, self-driving cars.

- Data Science
   - Data Science is an interdisciplinary field that involves collecting, cleaning, analyzing, and interpreting data to extract insights and support decision-making. It may use AI, ML, and DL as tools.
   - Example: Business analytics, forecasting sales, customer behavior analysis.
   - Scope: Interdisciplinary field focused on extracting insights from data for decision-making.
   - Techniques: Data collection, data cleaning, statistics, data visualization, ML/DL models.
   - Applications: Business intelligence, forecasting, market analysis, healthcare analytics, financial analysis

2. Explain overfitting and underfitting in ML. How can you detect and prevent them?

- In **machine learning**, two common problems that affect model performance are **underfitting** and **overfitting**.
  - **Underfitting** occurs when a model is too simple to capture the underlying pattern in the data. Such a model has **high bias**, meaning it makes strong assumptions about the data and ignores important relationships. As a result, it performs poorly on both the training data and unseen test data. Underfitting can be detected when both training error and validation error are high. It can be prevented by increasing model complexity, adding more relevant features, reducing excessive regularization, or training the model for more iterations.
  - **Overfitting** occurs when a model learns not only the underlying pattern but also the noise present in the training data. This happens due to **high variance**, where the model becomes overly sensitive to the training dataset. An overfitted model shows very low training error but high validation or test error. Overfitting can be detected by observing a large gap between training performance and test performance. It can be prevented by using more training data, reducing model complexity, applying regularization techniques, using cross-validation, and applying early stopping in iterative models such as neural networks.
 - The **bias–variance trade-off** explains the relationship between underfitting and overfitting. Bias refers to errors caused by overly simplistic models, leading to underfitting, while variance refers to errors caused by models that are too complex, leading to overfitting. A good machine learning model strikes a balance between bias and variance to achieve optimal generalization performance.
 - **Cross-validation** is a key technique used to detect both underfitting and overfitting. By dividing the dataset into multiple training and validation sets, cross-validation provides a reliable estimate of how well a model will perform on unseen data and helps in selecting appropriate model complexity and hyperparameters.

- **Regularization** techniques help prevent overfitting by adding a penalty to the loss function that discourages overly complex models. Common methods include L1 regularization (Lasso), which can shrink some coefficients to zero and perform feature selection; L2 regularization (Ridge), which penalizes large weights; Elastic Net, which combines L1 and L2; and dropout in deep learning, which randomly disables neurons during training to improve generalization.


3. How would you handle missing values in a dataset? Explain at least three methods with examples.

- Handling **missing values** is an important step in data preprocessing. The choice of method depends on the **amount of missing data**, the **nature of the variable**, and the **analysis goal**. Some common methods are explained below.

**1. Deletion Method**
In this method, observations (rows) or variables (columns) containing missing values are removed from the dataset.

* *Example:* If a dataset of 1,000 records has only 5 records with missing values, those rows can be safely deleted.
* *Advantage:* Simple to implement and does not introduce artificial data.
* *Limitation:* If many values are missing, deletion can reduce sample size and cause loss of important information.

**2. Mean / Median Imputation**
Missing values are replaced with the **mean or median** of the available values in that feature.

* *Example:* If the average age in a dataset is 35 years, missing age values can be replaced with 35. For skewed data, the median is preferred.
* *Advantage:* Easy and fast; maintains dataset size.
* *Limitation:* Reduces variability and may distort relationships between variables.

**3. Predictive Modeling (Model-Based Imputation)**
Missing values are predicted using **machine learning or statistical models** based on other features in the dataset.

* *Example:* Missing income values can be predicted using regression based on age, education, and occupation.
* *Advantage:* More accurate and preserves relationships among variables.
* *Limitation:* Computationally expensive and requires careful model selection.



4. What is an imbalanced dataset? Describe two techniques to handle it (theoretical + practical).

- An **imbalanced dataset** is a dataset in which the **class distribution is highly unequal**, meaning one class (majority class) has many more observations than the other class(es) (minority class). This is common in real-world problems such as fraud detection, disease diagnosis, and spam detection. In such cases, a model may become biased toward the majority class and achieve high accuracy while performing poorly on the minority class.

### **Techniques to Handle Imbalanced Datasets**

#### **1. Resampling Techniques (Data-Level Approach)**

**a) Random Oversampling**
This method increases the number of minority class samples by **randomly duplicating existing instances**.

* *Theoretical:* Balances the class distribution and reduces bias toward the majority class.
* *Practical Example:* In a fraud dataset where only 5% transactions are fraudulent, fraud samples are duplicated until they form a reasonable proportion.
* *Limitation:* May lead to overfitting due to repeated samples.

**b) Random Undersampling**
This method reduces the number of majority class samples by **randomly removing instances**.

* *Theoretical:* Simplifies the dataset and balances classes.
* *Practical Example:* From 95,000 non-fraud transactions, randomly keep only 10,000 to match minority class size.
* *Limitation:* Risk of losing important information.

#### **2. SMOTE (Synthetic Minority Over-sampling Technique)**

**SMOTE** generates **synthetic data points** for the minority class instead of duplicating existing ones.

* *Theoretical:* New samples are created by interpolating between a minority sample and its nearest neighbors.
* *Practical Example:* In a medical dataset, SMOTE creates synthetic patient records based on existing minority-class patients.
* *Advantage:* Reduces overfitting compared to random oversampling.
* *Limitation:* Can generate noisy or unrealistic samples if applied incorrectly.



#### **3. Class Weights (Algorithm-Level Approach)**

This method assigns **higher misclassification cost** to the minority class during model training.

* *Theoretical:* Forces the model to pay more attention to minority class errors.
* *Practical Example:* In logistic regression or SVM, setting `class_weight='balanced'` penalizes misclassification of rare events more heavily.
* *Advantage:* No change to data distribution.
* *Limitation:* Requires careful tuning.



5. Why is feature scaling important in ML? Compare Min-Max scaling and Standardization.

 - **Feature scaling** is the process of bringing all input features to a **similar scale** so that no single feature dominates the learning process due to its large numerical range. It is an essential preprocessing step in many machine learning algorithms.

- Feature scaling is especially important for **distance-based algorithms** such as **K-Nearest Neighbors (KNN)** and **Support Vector Machines (SVM)**. These algorithms rely on distance calculations; if one feature has values in thousands and another in decimals, the feature with larger values will dominate the distance measure, leading to biased and inaccurate results. Scaling ensures that all features contribute fairly to distance computations.

- Feature scaling also plays a crucial role in **gradient descent–based algorithms** such as linear regression, logistic regression, and neural networks. When features are on very different scales, the cost function becomes elongated, causing gradient descent to converge slowly or inconsistently. Scaling helps achieve faster and more stable convergence.


### **Min–Max Scaling**

Min–Max scaling rescales the data to a fixed range, usually between 0 and 1.

* **Characteristics:** Preserves the original distribution and relative distances.
* **Use Case:** Useful when features have a known bounded range.
* **Limitation:** Highly sensitive to outliers.


### **Standardization (Z-score Normalization)**

Standardization transforms data so that it has a **mean of 0** and a **standard deviation of 1**.

* **Characteristics:** Centers the data and reduces the impact of differing variances.
* **Use Case:** Preferred for algorithms like SVM, PCA, and gradient descent–based models.
* **Limitation:** Does not bound values to a fixed range.



6.  Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?

 - **Label Encoding** and **One-Hot Encoding** are techniques used to convert **categorical variables into numerical form** so that machine learning models can process them. The choice between them depends on whether the categorical variable has an **ordinal relationship** or is **nominal** in nature.

 - **Label Encoding** assigns a **unique integer value** to each category.
For example, a feature *Size* with categories {Small, Medium, Large} may be encoded as Small = 0, Medium = 1, Large = 2.
This method is suitable for **ordinal variables**, where the categories have a **natural order or ranking**. It preserves the order information and is computationally efficient. However, for nominal variables, label encoding can mislead models into assuming a false numerical relationship between categories.

- **One-Hot Encoding** creates **binary (0/1) columns** for each category.
For example, a feature *Color* with categories {Red, Blue, Green} is converted into three columns: Red, Blue, and Green, where each row has a value of 1 for its category and 0 for others.
This method is preferred for **nominal variables**, where there is **no inherent order** among categories. One-hot encoding prevents the model from assuming any ranking but increases dimensionality, especially when there are many categories.

 - **Preference Summary:**
Label encoding is preferred when the categorical variable is **ordinal** (e.g., education level, size, rating). One-hot encoding is preferred when the variable is **nominal** (e.g., gender, city, color). In practice, tree-based models can sometimes handle label-encoded nominal data better, while linear and distance-based models generally perform better with one-hot encoding.



7. Google Play Store Dataset a). Analyze the relationship between app categories and ratings. Which categories have the highest/lowest average ratings, and what could be the possible reasons? Dataset: https://github.com/MasteriNeuron/datasets.git

```python
import pandas as pd

# Load dataset directly from GitHub (RAW file)
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/master/googleplaystore.csv"
df = pd.read_csv(url)

# Display first 5 rows to confirm loading
print(df.head())

# Convert Rating column to numeric
df["Rating"] = pd.to_numeric(df["Rating"], errors="coerce")

# Remove missing values
df_clean = df.dropna(subset=["Category", "Rating"])

# Calculate average rating by category
avg_rating = (
    df_clean
    .groupby("Category")["Rating"]
    .mean()
    .sort_values(ascending=False)
)

print("Average Rating by Category:\n")
print(avg_rating)

print("\nTop 5 Categories with Highest Average Ratings:\n")
print(avg_rating.head(5))

print("\nBottom 5 Categories with Lowest Average Ratings:\n")
print(avg_rating.tail(5))
```

9.  Flight Price Prediction Dataset
- a) How do flight prices vary with the days left until departure? Identify any exponential price surges and recommend the best booking window.
- b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are consistently cheaper/premium, and why? Dataset: https://github.com/MasteriNeuron/datasets.git

```python
import pandas as pd
import numpy as np

# 1. Load dataset
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/flight_price.csv"
df = pd.read_csv(url)

# 2. Inspect columns (to avoid KeyError)
print("Columns in flight_price.csv:\n")
print(df.columns)

# 3. Print first few rows to understand structure
print("\nFirst 5 rows of dataset:\n")
print(df.head())

# Now identify the correct column names for price and days left
# For demonstration, let's assume:
# - 'Price' might be named 'price' or similar
# - 'Days_left' might be 'days_left' or similar
# You should replace below based on actual printed columns

# Attempt to find likely names (simple heuristic)
possible_price_cols = [c for c in df.columns if 'price' in c.lower()]
possible_days_cols = [c for c in df.columns if 'days' in c.lower()]

print("\nPossible price columns:", possible_price_cols)
print("Possible days left columns:", possible_days_cols)

# If none are found, inform user to check column names
if not possible_price_cols or not possible_days_cols:
    print("\n⚠️ Could not automatically detect price/days columns. Please check the printed column list.")
else:
    # Choose first matches
    price_col = possible_price_cols[0]
    days_col = possible_days_cols[0]

    # Convert to numeric
    df[price_col] = pd.to_numeric(df[price_col], errors="coerce")
    df[days_col] = pd.to_numeric(df[days_col], errors="coerce")

    # Clean data
    df_clean = df.dropna(subset=[price_col, days_col])

    # --- a) Price vs Days Left ---
    price_vs_days = df_clean.groupby(days_col)[price_col].mean().reset_index().sort_values(days_col, ascending=False)
    print("\nAverage price by days left:\n", price_vs_days.head(10))

    # Detect surges
    price_vs_days["pct_change"] = price_vs_days[price_col].pct_change() * 100
    surges = price_vs_days[price_vs_days["pct_change"] > 10]
    print("\nSignificant price surges (>10% increase):\n", surges)

    # --- b) Price comparison for Delhi–Mumbai ---
    if "Source" in df.columns and "Destination" in df.columns and "Airline" in df.columns:
        route_df = df_clean[(df_clean["Source"] == "Delhi") & (df_clean["Destination"] == "Mumbai")]
        avg_price_airlines = route_df.groupby("Airline")[price_col].mean().sort_values()
        print("\nAverage price by airline for Delhi–Mumbai:\n", avg_price_airlines)
    else:
        print("\n⚠️ Columns for route analysis ('Source', 'Destination', 'Airline') not found.")
```

10.  HR Analytics Dataset
- a). What factors most strongly correlate with employee attrition? Use visualizations to show key drivers (e.g., satisfaction, overtime, salary).
 - b). Are employees with more projects more likely to leave?  Dataset: hr_analytics

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load dataset
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/hr_analytics.csv"
df = pd.read_csv(url)

# 2. Inspect columns (so we can see the actual names)
print("Columns in the dataset:\n", df.columns)
print("\nFirst 5 rows:\n", df.head())

# 3. Try to identify the target/attrition-like column
possible_target_cols = [c for c in df.columns if "attr" in c.lower() or "left" in c.lower()]
print("\nPossible target/attrition columns:", possible_target_cols)

# Pick the first matching target column
if not possible_target_cols:
    raise ValueError("No obvious attrition/target column found. Please check df.columns")
target_col = possible_target_cols[0]

# Convert attrition target to binary if needed
if df[target_col].dtype == 'object':
    df["TargetFlag"] = df[target_col].apply(lambda x: 1 if str(x).lower() in ["yes", "y", "1", "true", "left", "left company"] else 0)
else:
    df["TargetFlag"] = df[target_col]

print(f"\nUsing '{target_col}' as target for attrition.")

# 4. Correlation for numeric features
numeric_cols = df.select_dtypes(include=["int64","float64"]).columns.tolist()
if "TargetFlag" not in numeric_cols:
    numeric_cols.append("TargetFlag")

corr = df[numeric_cols].corr()["TargetFlag"].sort_values(ascending=False)
print("\nCorrelation with attrition (numeric):\n", corr.head(10))

plt.figure(figsize=(10,5))
corr.drop("TargetFlag").plot(kind="bar", color="skyblue")
plt.title("Correlation of Numeric Features with Attrition")
plt.show()

# 5. Categorical drivers using simple plots
cat_cols = [c for c in df.columns if df[c].dtype == 'object' and c != target_col][:5]  # top 5 categorical features
for col in cat_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, hue="TargetFlag", data=df)
    plt.title(f"Attrition by {col}")
    plt.xticks(rotation=45)
    plt.show()

# 6. Project / workload effect (if available)
# Look for number of projects or similar columns
possible_proj_cols = [c for c in df.columns if "project" in c.lower() or "task" in c.lower()]
print("Possible project/workload columns:", possible_proj_cols)

for col in possible_proj_cols:
    plt.figure(figsize=(7,4))
    sns.boxplot(x="TargetFlag", y=col, data=df)
    plt.title(f"{col} vs Attrition")
    plt.show()
```
